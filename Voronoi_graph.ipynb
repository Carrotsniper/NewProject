```python
import numpy as np
from scipy.spatial import Voronoi, voronoi_plot_2d
import matplotlib.pyplot as plt

# 假设对于一个时间步的行人位置
points = np.array([[0, 0], [1, 0], [0, 1], [1, 1], [0.5, 0.5], [0, 0.1], [0.4,0.4]])

# 计算Voronoi图
vor = Voronoi(points)

# 绘制Voronoi图
fig = voronoi_plot_2d(vor)
plt.show()

# 构建邻接矩阵
n = len(points)
adj_matrix = np.zeros((n, n), dtype=int)
for point_idx, ridge_points in enumerate(vor.ridge_points):
    adj_matrix[ridge_points[0], ridge_points[1]] = 1
    adj_matrix[ridge_points[1], ridge_points[0]] = 1

print("邻接矩阵:")
print(adj_matrix)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import Delaunay

# 假设有一组点，这些点是你想进行三角剖分的对象
points = np.array([[0, 0], [1, 0], [0, 1], [1, 1], [0.5, 0.5]])

# 执行 Delaunay 三角剖分
tri = Delaunay(points)

# 绘制 Delaunay 三角剖分
plt.triplot(points[:, 0], points[:, 1], tri.simplices)
plt.plot(points[:, 0], points[:, 1], 'o')  # 将点标记出来
plt.show()

# 构建邻接矩阵
n = len(points)
adj_matrix = np.zeros((n, n), dtype=int)
for simplex in tri.simplices:
    for i in range(3):
        adj_matrix[simplex[i], simplex[(i + 1) % 3]] = 1
        adj_matrix[simplex[(i + 1) % 3], simplex[i]] = 1

print("邻接矩阵:")
print(adj_matrix)
```
